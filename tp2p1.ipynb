```python
import pandas as pd

dataset  = pd.read_csv('./ML/weather_numeric.csv')
```

```python
dataset
```

```python
dataset.describe()
```

```python
temps =['cool','temperate','hot']

dataset['temp_discret'] = pd.cut(x=dataset['temperature'], bins=[64,70.5,77.5,85], labels=temps)

dataset.head()
```

```python
humidts =['low','medium','high']

dataset['humdt_discret'] = pd.cut(x=dataset['humidity'], bins=[65,77.5,88,96], labels=humidts)

dataset.head()
```

```python
dataset.drop('temperature', axis=1, inplace=True)
```

```python
dataset.drop('humidity', axis=1, inplace=True)

dataset.head()
```

```python
dataset.shape
```

```python
dataset_lists = []
for i in range (0,14):
    dataset_lists.append([str(dataset.values[i,j]) for j in range(0,5)])
dataset_lists[:3]
```

```python
freq_item_set, rules = apriori(dataset_lists, minSup=0.2, minConf=0.5)
print(freq_item_set)
print(rules)
```
